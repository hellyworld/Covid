```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
confirmed=pd.read_csv('C:\\Users\\sf\\Desktop\\bi portfolio\\a3\\b\\covid19_confirmed.csv')
deaths=pd.read_csv('C:\\Users\\sf\\Desktop\\bi portfolio\\a3\\b\\covid19_deaths.csv')
recovered=pd.read_csv('C:\\Users\\sf\\Desktop\\bi portfolio\\a3\\b\\covid19_recovered.csv')
```

```python
confirmed=confirmed.drop(['Province_State', 'Lat', ], axis=1)
```

```python
deaths=deaths.drop(['Province_State'], axis=1)
```

```python
recovered=recovered.drop(['Province_State'], axis=1)
```

```python
confirmed=confirmed.groupby(confirmed['Country_Region']).aggregate('sum')
```

```python
confirmed.tail(50)
```

```python
deaths=deaths.groupby(deaths['Country_Region']).aggregate('sum')
```

```python
recovered=deaths.groupby(recovered['Country_Region']).aggregate('sum')
```

# New cases  

```python
new_cases=confirmed.copy()
```

```python
for day in range(1,len(confirmed)):
    new_cases.iloc[day]=confirmed.iloc[day]-confirmed.iloc[day-1]
print(new_cases.tail(50))
    
```

# growth rate of cases estimation 

```python
growth_rate=confirmed.copy()
```

```python
for day in range(1,len(confirmed)):
    growth_rate.iloc[day]=(new_cases.iloc[day] / confirmed.iloc[day-1]) *100
print(growth_rate)

print(growth_rate.tail(50))
  

    
```
